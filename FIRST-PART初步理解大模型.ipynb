### 1. 什么是大模型

属于深度神经网络，用于理解，生成和相应类似人类语言文本的神经网络，其特点就是大，具有数千亿个参数，被归类于生成式人工智能

### 2. 构建和使用的各个阶段

构建包括pre-training(预训练)和fine-tuning(微调)两个阶段，第一个是初始阶段，在大规模多样化的数据集中形成全面的语言理解能力，fine-tuning就是微调模型使其在特定任务和领域上进行针对训练，不同于传统机器学习中的监督学习，LLM采用的是自监督学习，并不需要人为为数据标签

### 3 .transformer架构


原始transformer是为了解决机器翻译问题而提出的，由编码器和解码器两部分组成，输入文本时，首先预处理文本生成特征向量等操作，进入编码器，生成用于解码的嵌入向量，作为解码器的输入，而翻译时是逐个单词实现，此时的部分输出即翻译后的内容同时作为输入到解码器中，不断迭代生成最终文本。这其中的关键就是自注意力机制，它让模型能够衡量句子中不同单词和词元的相对重要性，使得能够捕捉到长文本中的长距离依赖和上下文关系，提升其生成连贯上下文且上下文相关的输出的能力

为了适应不同的任务，transformer也有不同的变体，如BERT(基于transformer的双向编码器表示)和各种类GPT(Gernerative Pretrained Transformer)生成式预训练模型，BERT及其变体侧重于掩码预测，即预测出句子中缺失的部分，而GPT更擅长于帮你生成没说完的句子

### 4.GPT架构

GPT架构并不与原始的transformer架构相同，它只有解码器，将自己的输出作为未来预测的输入，是一种自回归模型，这大大提高了前后文本的一致性，也使得最初原定于解决机器翻译的GPT拥有了涌现的能力，即能够完成未经明确训练任务的能力，使得十分通用

### 5.学习构建大模型的顺序

第一阶段：数据准备与采样->理解注意力机制—>大语言模型架构—>预训练

第二阶段：训练循环->模型评估->加载预训练权重

第三阶段:微调以适应不同任务
